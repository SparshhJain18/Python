````
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import seaborn as sns

# Set style for better visualizations
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

def calculate_present_value(future_value, discount_rate, periods):
    """Calculate present value of a single future cash flow"""
    pv = future_value / (1 + discount_rate) ** periods
    return pv

def create_comprehensive_analysis(future_value, discount_rate, periods):
    """Create comprehensive present value analysis with visualizations"""

    # Calculate present value
    present_value = calculate_present_value(future_value, discount_rate, periods)

    print("🎮 MAYA'S CROWDFUNDING PRESENT VALUE ANALYSIS")
    print("=" * 60)
    print(f"Future Payment: ${future_value:,.2f}")
    print(f"Discount Rate: {discount_rate:.1%} per period")
    print(f"Time Periods: {periods}")
    print(f"Present Value: ${present_value:,.2f}")
    print()

    discount_amount = future_value - present_value
    discount_percentage = discount_amount / future_value

    print(f"💰 VALUE BREAKDOWN:")
    print(f"   Future Value:    ${future_value:,.2f}")
    print(f"   Present Value:   ${present_value:,.2f}")
    print(f"   Discount Amount: ${discount_amount:,.2f}")
    print(f"   Discount Rate:   {discount_percentage:.1%}")
    print()

    # Create the main visualization figure
    fig = plt.figure(figsize=(20, 16))

    # Create a grid layout for subplots
    gs = fig.add_gridspec(4, 3, height_ratios=[1, 1, 1, 1], width_ratios=[1, 1, 1])

    # 1. Present Value Timeline Visualization
    ax1 = fig.add_subplot(gs[0, :])

    # Timeline visualization
    timeline_periods = np.arange(0, periods + 1)
    timeline_values = np.zeros(len(timeline_periods))
    timeline_values[0] = present_value  # Present value at time 0
    timeline_values[-1] = future_value  # Future value at final period

    # Create timeline
    ax1.plot([0, periods], [0, 0], 'k-', linewidth=3, alpha=0.3)

    # Add time markers
    for i in timeline_periods:
        ax1.plot([i, i], [-future_value*0.05, future_value*0.05], 'k-', linewidth=2)
        ax1.text(i, -future_value*0.1, f'Period {i}', ha='center', va='top', fontsize=10)

    # Present value arrow and annotation
    ax1.annotate(f'Present Value\n${present_value:,.0f}',
                xy=(0, present_value*0.8), xytext=(0, present_value*1.2),
                ha='center', va='bottom', fontsize=14, fontweight='bold',
                arrowprops=dict(arrowstyle='->', color='green', lw=3),
                bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgreen", alpha=0.8))

    # Future value arrow and annotation
    ax1.annotate(f'Future Value\n${future_value:,.0f}',
                xy=(periods, future_value*0.8), xytext=(periods, future_value*1.2),
                ha='center', va='bottom', fontsize=14, fontweight='bold',
                arrowprops=dict(arrowstyle='->', color='red', lw=3),
                bbox=dict(boxstyle="round,pad=0.3", facecolor="lightcoral", alpha=0.8))

    # Growth curve
    time_curve = np.linspace(0, periods, 100)
    value_curve = present_value * (1 + discount_rate) ** time_curve
    ax1.plot(time_curve, value_curve, 'b--', linewidth=3, alpha=0.7, label=f'Growth at {discount_rate:.1%}')

    ax1.set_xlim(-0.5, periods + 0.5)
    ax1.set_ylim(-future_value*0.2, future_value*1.4)
    ax1.set_title(f'Present Value Timeline: ${present_value:,.0f} grows to ${future_value:,.0f} in {periods} periods',
                  fontsize=16, fontweight='bold')
    ax1.legend(fontsize=12)
    ax1.grid(True, alpha=0.3)

    # 2. Discount Factor Visualization
    ax2 = fig.add_subplot(gs[1, 0])

    discount_factor = 1 / (1 + discount_rate) ** periods
    periods_range = np.arange(1, periods + 1)
    discount_factors = 1 / (1 + discount_rate) ** periods_range

    bars = ax2.bar(periods_range, discount_factors, alpha=0.7, color='skyblue', edgecolor='navy')
    ax2.axhline(y=discount_factor, color='red', linestyle='--', linewidth=2,
                label=f'Our Factor: {discount_factor:.4f}')

    # Highlight our specific period
    bars[periods-1].set_color('red')
    bars[periods-1].set_alpha(0.8)

    ax2.set_xlabel('Periods')
    ax2.set_ylabel('Discount Factor')
    ax2.set_title(f'Discount Factors at {discount_rate:.1%} Rate')
    ax2.legend()
    ax2.grid(True, alpha=0.3)

    # 3. Value Breakdown Pie Chart
    ax3 = fig.add_subplot(gs[1, 1])

    sizes = [present_value, discount_amount]
    labels = [f'Present Value\n${present_value:,.0f}', f'Time Value\n${discount_amount:,.0f}']
    colors = ['lightgreen', 'lightcoral']
    explode = (0.1, 0)

    wedges, texts, autotexts = ax3.pie(sizes, explode=explode, labels=labels, colors=colors,
                                      autopct='%1.1f%%', shadow=True, startangle=90,
                                      textprops={'fontsize': 12})

    ax3.set_title('Future Value Composition', fontsize=14, fontweight='bold')

    # 4. Sensitivity Analysis - Rate Impact
    ax4 = fig.add_subplot(gs[1, 2])

    rate_range = np.linspace(0.01, 0.15, 50)
    pv_sensitivity = future_value / (1 + rate_range) ** periods

    ax4.plot(rate_range * 100, pv_sensitivity, 'b-', linewidth=3, label='Present Value')
    ax4.axvline(x=discount_rate * 100, color='red', linestyle='--', linewidth=2,
                label=f'Current Rate: {discount_rate:.1%}')
    ax4.axhline(y=present_value, color='red', linestyle='--', linewidth=2, alpha=0.7)
    ax4.scatter([discount_rate * 100], [present_value], color='red', s=100, zorder=5)

    ax4.set_xlabel('Discount Rate (%)')
    ax4.set_ylabel('Present Value ($)')
    ax4.set_title('PV Sensitivity to Discount Rate')
    ax4.legend()
    ax4.grid(True, alpha=0.3)

    # 5. Time Sensitivity Analysis
    ax5 = fig.add_subplot(gs[2, 0])

    time_range = np.arange(1, 21)
    pv_time_sensitivity = future_value / (1 + discount_rate) ** time_range

    ax5.plot(time_range, pv_time_sensitivity, 'g-', linewidth=3, marker='o', markersize=4)
    ax5.axvline(x=periods, color='red', linestyle='--', linewidth=2,
                label=f'Current Periods: {periods}')
    ax5.axhline(y=present_value, color='red', linestyle='--', linewidth=2, alpha=0.7)
    ax5.scatter([periods], [present_value], color='red', s=100, zorder=5)

    ax5.set_xlabel('Time Periods')
    ax5.set_ylabel('Present Value ($)')
    ax5.set_title('PV Sensitivity to Time')
    ax5.legend()
    ax5.grid(True, alpha=0.3)

    # 6. Compound Growth Visualization
    ax6 = fig.add_subplot(gs[2, 1])

    growth_periods = np.arange(0, periods + 1)
    growth_values = present_value * (1 + discount_rate) ** growth_periods

    ax6.bar(growth_periods, growth_values, alpha=0.7, color='gold', edgecolor='orange')
    ax6.plot(growth_periods, growth_values, 'ro-', linewidth=2, markersize=6)

    # Add value labels on bars
    for i, v in enumerate(growth_values):
        ax6.text(i, v + future_value*0.05, f'${v:,.0f}', ha='center', va='bottom',
                fontweight='bold', fontsize=10)

    ax6.set_xlabel('Period')
    ax6.set_ylabel('Value ($)')
    ax6.set_title(f'Growth from ${present_value:,.0f} at {discount_rate:.1%}')
    ax6.grid(True, alpha=0.3)

    # 7. Decision Matrix Heatmap
    ax7 = fig.add_subplot(gs[2, 2])

    # Create decision matrix for different rates and periods
    rate_matrix = np.linspace(0.02, 0.10, 9)
    period_matrix = np.arange(1, 10)
    pv_matrix = np.zeros((len(period_matrix), len(rate_matrix)))

    for i, p in enumerate(period_matrix):
        for j, r in enumerate(rate_matrix):
            pv_matrix[i, j] = future_value / (1 + r) ** p

    im = ax7.imshow(pv_matrix, cmap='RdYlGn', aspect='auto')
    ax7.set_xticks(range(len(rate_matrix)))
    ax7.set_xticklabels([f'{r:.1%}' for r in rate_matrix])
    ax7.set_yticks(range(len(period_matrix)))
    ax7.set_yticklabels(period_matrix)
    ax7.set_xlabel('Discount Rate')
    ax7.set_ylabel('Periods')
    ax7.set_title('Present Value Heatmap')

    # Add our current scenario marker
    current_rate_idx = np.argmin(np.abs(rate_matrix - discount_rate))
    current_period_idx = periods - 1
    ax7.plot(current_rate_idx, current_period_idx, 'r*', markersize=20, markeredgecolor='black')

    plt.colorbar(im, ax=ax7, label='Present Value ($)')

    # 8. Risk Analysis
    ax8 = fig.add_subplot(gs[3, :])

    # Create risk scenarios
    scenarios = ['Pessimistic', 'Expected', 'Optimistic']
    rate_scenarios = [discount_rate + 0.02, discount_rate, discount_rate - 0.02]
    pv_scenarios = [future_value / (1 + r) ** periods for r in rate_scenarios]

    colors_scenario = ['red', 'blue', 'green']
    bars = ax8.bar(scenarios, pv_scenarios, color=colors_scenario, alpha=0.7, edgecolor='black')

    # Add value labels
    for bar, pv, rate in zip(bars, pv_scenarios, rate_scenarios):
        height = bar.get_height()
        ax8.text(bar.get_x() + bar.get_width()/2., height + future_value*0.02,
                f'${pv:,.0f}\n({rate:.1%})', ha='center', va='bottom',
                fontweight='bold', fontsize=12)

    ax8.axhline(y=present_value, color='blue', linestyle='--', linewidth=2, alpha=0.7,
                label=f'Expected PV: ${present_value:,.0f}')
    ax8.set_ylabel('Present Value ($)')
    ax8.set_title('Risk Scenario Analysis (±2% rate variation)')
    ax8.legend()
    ax8.grid(True, alpha=0.3, axis='y')

    plt.tight_layout()
    plt.show()

    # Additional Analysis Table
    print("📊 DETAILED ANALYSIS TABLE:")
    print("=" * 50)

    analysis_data = {
        'Metric': [
            'Future Value', 'Present Value', 'Discount Factor', 'Time Value Lost',
            'Effective Annual Loss', 'Break-even Rate', 'Years to Double'
        ],
        'Value': [
            f"${future_value:,.2f}",
            f"${present_value:,.2f}",
            f"{discount_factor:.6f}",
            f"${discount_amount:,.2f}",
            f"{discount_percentage/periods:.2%}",
            f"{discount_rate:.2%}",
            f"{np.log(2)/np.log(1+discount_rate):.1f} periods"
        ],
        'Interpretation': [
            'Amount to be received in future',
            'Worth in today\'s dollars',
            'Multiplier to get present value',
            'Cost of waiting for money',
            'Value lost per period on average',
            'Rate that makes PV = 0',
            'Time for money to double at this rate'
        ]
    }

    analysis_df = pd.DataFrame(analysis_data)
    print(analysis_df.to_string(index=False))
    print()

    return present_value

def main():
    """Main function to get user inputs and run analysis"""
    print("🎮 MAYA'S CROWDFUNDING PRESENT VALUE CALCULATOR")
    print("=" * 60)
    print("Enter the parameters for your crowdfunding payment analysis:")
    print()

    try:
        # Get user inputs
        future_value = float(input("💰 Enter the future payment amount ($): $"))
        discount_rate = float(input("📈 Enter the discount rate per period (as decimal, e.g., 0.065 for 6.5%): "))
        periods = int(input("⏰ Enter the number of periods until payment: "))

        print("\n" + "="*60)
        print("CALCULATING PRESENT VALUE...")
        print("="*60)

        # Run comprehensive analysis
        present_value = create_comprehensive_analysis(future_value, discount_rate, periods)

        print("\n🎯 FINAL RECOMMENDATION:")
        print("="*30)
        if present_value > future_value * 0.7:  # If PV is more than 70% of FV
            print("✅ GOOD DEAL: The present value is relatively high")
            print("   The time value loss is reasonable for this investment")
        elif present_value > future_value * 0.5:  # If PV is 50-70% of FV
            print("⚠️  MODERATE DEAL: Significant time value loss")
            print("   Consider if the return justifies the wait")
        else:  # If PV is less than 50% of FV
            print("❌ POOR DEAL: Substantial time value erosion")
            print("   The long wait significantly reduces the payment's value")

        print(f"\n💡 INSIGHT: You're effectively 'paying' ${future_value - present_value:,.2f}")
        print(f"   in opportunity cost for waiting {periods} periods.")

    except ValueError:
        print("❌ Error: Please enter valid numbers for all inputs.")
        return
    except KeyboardInterrupt:
        print("\n\n👋 Analysis cancelled by user.")
        return

# Run the main function
if __name__ == "__main__":
    main()
````

````
#Scenario 4
import matplotlib.pyplot as plt
import numpy as np

# Scenario 4: The Sound Design Savings Fund
# Maya's annuity calculation for sound design hiring fund

# Given data
monthly_deposit = 75.00  # Deposit at end of each period
interest_rate = 0.085  # 8.5% per period
periods = 5  # Number of deposits (periods 1-5)
final_period = 6  # Final evaluation period

print("=" * 60)
print("MAYA'S SOUND DESIGN SAVINGS FUND ANALYSIS")
print("=" * 60)
print(f"Monthly Deposit: ${monthly_deposit:,.2f}")
print(f"Interest Rate per Period: {interest_rate:.1%}")
print(f"Deposit Periods: 1 through {periods}")
print(f"Final Evaluation: End of Period {final_period}")
print("-" * 60)

# Calculate future value of annuity
# Since deposits are made at the END of each period (ordinary annuity)
# and we want value at end of period 6, each deposit earns interest for:
# Period 1 deposit: earns interest for 5 periods (periods 2,3,4,5,6)
# Period 2 deposit: earns interest for 4 periods (periods 3,4,5,6)
# Period 3 deposit: earns interest for 3 periods (periods 4,5,6)
# Period 4 deposit: earns interest for 2 periods (periods 5,6)
# Period 5 deposit: earns interest for 1 period (period 6)

# Method 1: Calculate each deposit's future value individually
deposit_values = []
periods_list = []
future_values = []

total_future_value = 0

print("\nDETAILED BREAKDOWN:")
print("=" * 40)
for deposit_period in range(1, periods + 1):
    # Calculate how many periods this deposit will earn interest
    interest_periods = final_period - deposit_period
    # Future value of this specific deposit
    fv_deposit = monthly_deposit * (1 + interest_rate) ** interest_periods
    total_future_value += fv_deposit

    deposit_values.append(fv_deposit)
    periods_list.append(deposit_period)
    future_values.append(total_future_value)

    print(f"Deposit in Period {deposit_period}: ${monthly_deposit:,.2f} → "
          f"${fv_deposit:,.2f} (grows for {interest_periods} periods)")

print("-" * 40)
print(f"TOTAL FUTURE VALUE AT END OF PERIOD 6: ${total_future_value:,.2f}")

# Method 2: Using standard annuity formula for verification
# FV = PMT × [((1+r)^n - 1) / r] × (1+r)
# The (1+r) factor accounts for one additional period of growth since
# we want value at end of period 6, not end of period 5
fv_formula = monthly_deposit * (((1 + interest_rate)**periods - 1) / interest_rate) * (1 + interest_rate)
print(f"Verification using annuity formula: ${fv_formula:,.2f}")

# Calculate cumulative account balance over time
account_balance = [0]  # Start with $0 at period 0
period_labels = [0]

for period in range(1, final_period + 1):
    if period <= periods:  # Making deposits in periods 1-5
        # Previous balance grows with interest, then add new deposit
        new_balance = account_balance[-1] * (1 + interest_rate) + monthly_deposit
    else:  # Period 6: no deposit, just growth
        new_balance = account_balance[-1] * (1 + interest_rate)

    account_balance.append(new_balance)
    period_labels.append(period)

# Create visualizations
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))
fig.suptitle('Maya\'s Sound Design Savings Fund Analysis', fontsize=16, fontweight='bold')

# Graph 1: Individual Deposit Future Values
ax1.bar(periods_list, deposit_values, color='skyblue', alpha=0.8, edgecolor='navy')
ax1.set_xlabel('Deposit Period')
ax1.set_ylabel('Future Value ($)')
ax1.set_title('Future Value of Each Individual Deposit at End of Period 6')
ax1.grid(True, alpha=0.3)

# Add value labels on bars
for i, (period, value) in enumerate(zip(periods_list, deposit_values)):
    ax1.text(period, value + 2, f'${value:.2f}', ha='center', va='bottom', fontweight='bold')

ax1.set_xticks(periods_list)
ax1.set_ylim(0, max(deposit_values) * 1.15)

# Graph 2: Cumulative Account Balance Over Time
ax2.plot(period_labels, account_balance, marker='o', linewidth=3, markersize=8,
         color='green', markerfacecolor='lightgreen', markeredgecolor='darkgreen')
ax2.fill_between(period_labels, account_balance, alpha=0.3, color='green')
ax2.set_xlabel('Period')
ax2.set_ylabel('Account Balance ($)')
ax2.set_title('Cumulative Account Balance Over Time')
ax2.grid(True, alpha=0.3)

# Add value labels on points
for period, balance in zip(period_labels, account_balance):
    if period == 0:
        ax2.text(period, balance + 10, f'${balance:.0f}', ha='center', va='bottom', fontweight='bold')
    else:
        ax2.text(period, balance + 10, f'${balance:.2f}', ha='center', va='bottom', fontweight='bold')

# Highlight deposits with arrows
for period in range(1, periods + 1):
    ax2.annotate(f'Deposit\n${monthly_deposit}',
                xy=(period, account_balance[period]),
                xytext=(period + 0.3, account_balance[period] - 30),
                arrowprops=dict(arrowstyle='->', color='red', lw=2),
                fontsize=9, ha='center', color='red', fontweight='bold')

ax2.set_xticks(period_labels)
ax2.set_ylim(0, max(account_balance) * 1.2)

plt.tight_layout()
plt.show()

# Summary statistics
total_deposits = monthly_deposit * periods
interest_earned = total_future_value - total_deposits

print("\n" + "=" * 60)
print("SUMMARY STATISTICS")
print("=" * 60)
print(f"Total Deposits Made: ${total_deposits:,.2f}")
print(f"Interest Earned: ${interest_earned:,.2f}")
print(f"Final Account Value: ${total_future_value:,.2f}")
print(f"Effective Return: {(interest_earned/total_deposits)*100:.2f}%")
print("=" * 60)
````

````
# scenario 5 and 6
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

print("MAYA'S GAME STUDIO: DLC PROJECT ANALYSIS")
print("=" * 50)

# SCENARIO 5: "Side Quest" DLC Project
print("SCENARIO 5: 'SIDE QUEST' DLC PROJECT")
print("=" * 45)

# Side Quest project parameters
sq_initial_investment = 80.00  # thousand dollars
sq_discount_rate = 0.07  # 7.0%
sq_cash_flows = np.array([55.00, 85.00, 110.00, 90.00, 70.00])  # thousands
sq_periods = np.arange(1, len(sq_cash_flows) + 1)

print(f"Initial Investment: ${sq_initial_investment:.2f}k")
print(f"Discount Rate: {sq_discount_rate:.1%}")
print("Projected Cash Flows (in thousands):")
for i, cf in enumerate(sq_cash_flows, 1):
    print(f"  Period {i}: ${cf:.2f}k")
print()

# Calculate NPV for Side Quest
sq_pv_factors = (1 + sq_discount_rate) ** (-sq_periods)
sq_pv_cash_flows = sq_cash_flows * sq_pv_factors
sq_total_pv = np.sum(sq_pv_cash_flows)
sq_npv = sq_total_pv - sq_initial_investment

print("Side Quest NPV Calculation:")
print("-" * 30)
sq_analysis_data = {
    'Period': sq_periods,
    'Cash Flow ($k)': sq_cash_flows,
    'PV Factor': sq_pv_factors,
    'Present Value ($k)': sq_pv_cash_flows
}

sq_df = pd.DataFrame(sq_analysis_data)
print(sq_df.to_string(index=False, float_format='%.4f'))
print()
print(f"Total Present Value of Cash Flows: ${sq_total_pv:.2f}k")
print(f"Initial Investment: ${sq_initial_investment:.2f}k")
print(f"NET PRESENT VALUE (NPV): ${sq_npv:.2f}k")
print()

if sq_npv > 0:
    print("✅ RECOMMENDATION: ACCEPT - Project creates value")
else:
    print("❌ RECOMMENDATION: REJECT - Project destroys value")
print()

# SCENARIO 6: "Cross-Platform" Project with Inflation
print("SCENARIO 6: 'CROSS-PLATFORM' PROJECT (WITH INFLATION)")
print("=" * 60)

# Cross-Platform project parameters
cp_initial_investment = 140.00  # thousand dollars
cp_nominal_cash_flows = np.array([45.00, 55.00, 80.00, 105.00, 60.00])  # thousands (nominal)
cp_inflation_rate = 0.031  # 3.1%
cp_nominal_discount_rate = 0.09  # 9.0%
cp_periods = np.arange(1, len(cp_nominal_cash_flows) + 1)

print(f"Initial Investment: ${cp_initial_investment:.2f}k")
print(f"Inflation Rate: {cp_inflation_rate:.1%}")
print(f"Nominal Discount Rate: {cp_nominal_discount_rate:.1%}")
print()

# Calculate real discount rate
# (1 + nominal rate) = (1 + real rate) × (1 + inflation rate)
# Real rate = (1 + nominal rate) / (1 + inflation rate) - 1
cp_real_discount_rate = (1 + cp_nominal_discount_rate) / (1 + cp_inflation_rate) - 1

print(f"Real Discount Rate: {cp_real_discount_rate:.2%}")
print()

print("Projected Cash Flows (in thousands):")
for i, cf in enumerate(cp_nominal_cash_flows, 1):
    print(f"  Period {i}: ${cf:.2f}k (nominal)")
print()

# Method 1: Convert nominal cash flows to real, then discount at real rate
print("METHOD 1: Real Cash Flows with Real Discount Rate")
print("-" * 50)

# Convert nominal cash flows to real (today's purchasing power)
inflation_factors = (1 + cp_inflation_rate) ** cp_periods
cp_real_cash_flows = cp_nominal_cash_flows / inflation_factors

# Discount real cash flows at real rate
cp_real_pv_factors = (1 + cp_real_discount_rate) ** (-cp_periods)
cp_real_pv_cash_flows = cp_real_cash_flows * cp_real_pv_factors
cp_real_total_pv = np.sum(cp_real_pv_cash_flows)
cp_real_npv = cp_real_total_pv - cp_initial_investment

real_analysis_data = {
    'Period': cp_periods,
    'Nominal CF ($k)': cp_nominal_cash_flows,
    'Inflation Factor': inflation_factors,
    'Real CF ($k)': cp_real_cash_flows,
    'PV Factor (Real)': cp_real_pv_factors,
    'Present Value ($k)': cp_real_pv_cash_flows
}

real_df = pd.DataFrame(real_analysis_data)
print(real_df.to_string(index=False, float_format='%.4f'))
print()
print(f"Total Present Value (Real Method): ${cp_real_total_pv:.2f}k")
print(f"Initial Investment: ${cp_initial_investment:.2f}k")
print(f"NPV (Real Method): ${cp_real_npv:.2f}k")
print()

# Method 2: Discount nominal cash flows at nominal rate (verification)
print("METHOD 2: Nominal Cash Flows with Nominal Discount Rate")
print("-" * 55)

cp_nominal_pv_factors = (1 + cp_nominal_discount_rate) ** (-cp_periods)
cp_nominal_pv_cash_flows = cp_nominal_cash_flows * cp_nominal_pv_factors
cp_nominal_total_pv = np.sum(cp_nominal_pv_cash_flows)
cp_nominal_npv = cp_nominal_total_pv - cp_initial_investment

nominal_analysis_data = {
    'Period': cp_periods,
    'Nominal CF ($k)': cp_nominal_cash_flows,
    'PV Factor (Nominal)': cp_nominal_pv_factors,
    'Present Value ($k)': cp_nominal_pv_cash_flows
}

nominal_df = pd.DataFrame(nominal_analysis_data)
print(nominal_df.to_string(index=False, float_format='%.4f'))
print()
print(f"Total Present Value (Nominal Method): ${cp_nominal_total_pv:.2f}k")
print(f"Initial Investment: ${cp_initial_investment:.2f}k")
print(f"NPV (Nominal Method): ${cp_nominal_npv:.2f}k")
print()

# Verification that both methods give same result
npv_difference = abs(cp_real_npv - cp_nominal_npv)
print(f"NPV Difference between methods: ${npv_difference:.6f}k")
print("✓ Both methods should give identical results (within rounding)")
print()

if cp_real_npv > 0:
    print("✅ RECOMMENDATION: ACCEPT - Project creates value")
else:
    print("❌ RECOMMENDATION: REJECT - Project destroys value")
print()

# PROJECT COMPARISON
print("PROJECT COMPARISON ANALYSIS")
print("=" * 35)

comparison_data = {
    'Project': ['Side Quest DLC', 'Cross-Platform DLC'],
    'Initial Investment ($k)': [sq_initial_investment, cp_initial_investment],
    'NPV ($k)': [sq_npv, cp_real_npv],
    'Investment Multiple': [sq_npv/sq_initial_investment, cp_real_npv/cp_initial_investment],
    'Recommendation': [
        'Accept' if sq_npv > 0 else 'Reject',
        'Accept' if cp_real_npv > 0 else 'Reject'
    ]
}

comparison_df = pd.DataFrame(comparison_data)
print(comparison_df.to_string(index=False))
print()

# Determine best project
if sq_npv > 0 and cp_real_npv > 0:
    if sq_npv > cp_real_npv:
        best_project = "Side Quest DLC"
        best_npv = sq_npv
    else:
        best_project = "Cross-Platform DLC"
        best_npv = cp_real_npv
    print(f"🏆 BEST PROJECT: {best_project} (NPV: ${best_npv:.2f}k)")
elif sq_npv > 0:
    print("🏆 BEST PROJECT: Side Quest DLC (only positive NPV project)")
elif cp_real_npv > 0:
    print("🏆 BEST PROJECT: Cross-Platform DLC (only positive NPV project)")
else:
    print("⚠️  NEITHER PROJECT should be undertaken (both have negative NPV)")
print()

# SENSITIVITY ANALYSIS
print("SENSITIVITY ANALYSIS")
print("=" * 25)

# Test different discount rates for Side Quest
discount_rates = np.linspace(0.03, 0.12, 20)
sq_npvs = []
cp_npvs = []

for rate in discount_rates:
    # Side Quest sensitivity
    pv_factors = (1 + rate) ** (-sq_periods)
    pv_cash_flows = sq_cash_flows * pv_factors
    total_pv = np.sum(pv_cash_flows)
    npv = total_pv - sq_initial_investment
    sq_npvs.append(npv)

    # Cross-Platform sensitivity (using real approach)
    real_rate = (1 + rate) / (1 + cp_inflation_rate) - 1
    real_pv_factors = (1 + real_rate) ** (-cp_periods)
    real_pv_cash_flows = cp_real_cash_flows * real_pv_factors
    real_total_pv = np.sum(real_pv_cash_flows)
    real_npv = real_total_pv - cp_initial_investment
    cp_npvs.append(real_npv)

# Find break-even rates (IRR approximation)
sq_breakeven_idx = np.argmin(np.abs(sq_npvs))
cp_breakeven_idx = np.argmin(np.abs(cp_npvs))
sq_irr_approx = discount_rates[sq_breakeven_idx]
cp_irr_approx = discount_rates[cp_breakeven_idx]

print(f"Approximate IRR - Side Quest: {sq_irr_approx:.2%}")
print(f"Approximate IRR - Cross-Platform: {cp_irr_approx:.2%}")
print()

# VISUALIZATIONS
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))

# Plot 1: Cash Flow Timeline - Side Quest
ax1.bar(sq_periods, sq_cash_flows, alpha=0.7, color='lightblue', edgecolor='navy', label='Cash Flows')
ax1.axhline(y=0, color='black', linestyle='-', alpha=0.3)
ax1.bar([0], [-sq_initial_investment], alpha=0.7, color='red', edgecolor='darkred', label='Initial Investment')
ax1.set_xlabel('Period')
ax1.set_ylabel('Cash Flow ($k)')
ax1.set_title('Side Quest DLC - Cash Flow Timeline')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Add NPV annotation
ax1.text(3, 120, f'NPV: ${sq_npv:.2f}k', fontsize=12, fontweight='bold',
         bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgreen" if sq_npv > 0 else "lightcoral"))

# Plot 2: Cash Flow Timeline - Cross-Platform (Real vs Nominal)
width = 0.35
x = np.arange(len(cp_periods))
ax2.bar(x - width/2, cp_nominal_cash_flows, width, label='Nominal Cash Flows', alpha=0.7, color='orange')
ax2.bar(x + width/2, cp_real_cash_flows, width, label='Real Cash Flows', alpha=0.7, color='green')
ax2.bar([-0.5], [-cp_initial_investment], width, alpha=0.7, color='red', label='Initial Investment')
ax2.set_xlabel('Period')
ax2.set_ylabel('Cash Flow ($k)')
ax2.set_title('Cross-Platform DLC - Nominal vs Real Cash Flows')
ax2.set_xticks(range(len(cp_periods)))
ax2.set_xticklabels(cp_periods)
ax2.legend()
ax2.grid(True, alpha=0.3)

# Add NPV annotation
ax2.text(2, 110, f'NPV: ${cp_real_npv:.2f}k', fontsize=12, fontweight='bold',
         bbox=dict(boxstyle="round,pad=0.3", facecolor="lightgreen" if cp_real_npv > 0 else "lightcoral"))

# Plot 3: NPV Sensitivity Analysis
ax3.plot(discount_rates * 100, sq_npvs, 'b-', linewidth=2, label='Side Quest', marker='o', markersize=4)
ax3.plot(discount_rates * 100, cp_npvs, 'r-', linewidth=2, label='Cross-Platform', marker='s', markersize=4)
ax3.axhline(y=0, color='black', linestyle='--', alpha=0.7, label='Break-even')
ax3.axvline(x=sq_discount_rate * 100, color='blue', linestyle=':', alpha=0.7, label='Side Quest Rate (7%)')
ax3.axvline(x=cp_nominal_discount_rate * 100, color='red', linestyle=':', alpha=0.7, label='Cross-Platform Rate (9%)')
ax3.set_xlabel('Discount Rate (%)')
ax3.set_ylabel('NPV ($k)')
ax3.set_title('NPV Sensitivity to Discount Rate')
ax3.legend()
ax3.grid(True, alpha=0.3)

# Plot 4: Present Value Breakdown
categories = ['Side Quest\nTotal PV', 'Side Quest\nInvestment', 'Cross-Platform\nTotal PV', 'Cross-Platform\nInvestment']
values = [sq_total_pv, sq_initial_investment, cp_real_total_pv, cp_initial_investment]
colors = ['lightblue', 'lightcoral', 'lightgreen', 'lightsalmon']

bars = ax4.bar(categories, values, color=colors, alpha=0.7, edgecolor='black')
ax4.set_ylabel('Amount ($k)')
ax4.set_title('Present Value vs Investment Comparison')
ax4.grid(True, alpha=0.3, axis='y')

# Add value labels on bars
for bar, value in zip(bars, values):
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height + 2,
             f'${value:.1f}k', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# INFLATION IMPACT ANALYSIS
print("INFLATION IMPACT ANALYSIS")
print("=" * 30)
print("Effect of inflation on Cross-Platform project:")
print()

inflation_scenarios = [0.00, 0.02, 0.031, 0.05, 0.08]
for inflation in inflation_scenarios:
    real_rate = (1 + cp_nominal_discount_rate) / (1 + inflation) - 1
    inflation_factors_scenario = (1 + inflation) ** cp_periods
    real_cash_flows_scenario = cp_nominal_cash_flows / inflation_factors_scenario
    real_pv_factors_scenario = (1 + real_rate) ** (-cp_periods)
    real_pv_scenario = np.sum(real_cash_flows_scenario * real_pv_factors_scenario)
    npv_scenario = real_pv_scenario - cp_initial_investment

    print(f"Inflation Rate: {inflation:.1%} → Real Rate: {real_rate:.2%} → NPV: ${npv_scenario:.2f}k")

print()
print("→ INSIGHT: Higher inflation reduces the real value of future cash flows")
print("→ INSIGHT: This makes long-term projects riskier in inflationary environments")
print()

print("FINAL RECOMMENDATIONS:")
print("=" * 25)
print("1. FINANCIAL ANALYSIS:")
if sq_npv > 0 and cp_real_npv > 0:
    print(f"   Both projects create value, but {best_project} is superior")
elif sq_npv > 0:
    print("   Only Side Quest DLC should be undertaken")
elif cp_real_npv > 0:
    print("   Only Cross-Platform DLC should be undertaken")
else:
    print("   Neither project meets the required return threshold")

print()
print("2. RISK CONSIDERATIONS:")
print("   • Side Quest: Lower investment, shorter payback period")
print("   • Cross-Platform: Higher investment, inflation-sensitive cash flows")
print()
print("3. STRATEGIC FACTORS:")
print("   • Market expansion potential of Cross-Platform")
print("   • Resource requirements and opportunity costs")
print("   • Portfolio diversification benefits")
````

````
#scenario 9
import matplotlib.pyplot as plt
import numpy as np

# Mortgage parameters
loan_amount = 500000  # $500,000 loan
years = 30  # 30-year mortgage
months = years * 12  # Total number of payments

# Interest rate scenarios
rate_high = 0.095  # 9.5% initial rate
rate_low = 0.082   # 8.2% dropped rate

def calculate_monthly_payment(principal, annual_rate, years):
    """Calculate monthly mortgage payment using the standard formula"""
    monthly_rate = annual_rate / 12
    num_payments = years * 12

    # Monthly payment formula: P * [r(1+r)^n] / [(1+r)^n - 1]
    monthly_payment = principal * (monthly_rate * (1 + monthly_rate)**num_payments) / ((1 + monthly_rate)**num_payments - 1)
    return monthly_payment

def create_amortization_schedule(principal, annual_rate, years):
    """Create complete amortization schedule"""
    monthly_rate = annual_rate / 12
    num_payments = years * 12
    monthly_payment = calculate_monthly_payment(principal, annual_rate, years)

    remaining_balance = principal
    schedule = []

    for month in range(1, num_payments + 1):
        interest_payment = remaining_balance * monthly_rate
        principal_payment = monthly_payment - interest_payment
        remaining_balance -= principal_payment

        # Handle final payment rounding
        if remaining_balance < 0.01:
            principal_payment += remaining_balance
            remaining_balance = 0

        schedule.append({
            'month': month,
            'year': (month - 1) // 12 + 1,
            'payment': monthly_payment,
            'principal': principal_payment,
            'interest': interest_payment,
            'balance': max(0, remaining_balance)
        })

    return schedule

# Calculate payments for both rates
payment_high = calculate_monthly_payment(loan_amount, rate_high, years)
payment_low = calculate_monthly_payment(loan_amount, rate_low, years)

total_high = payment_high * months
total_low = payment_low * months

interest_high = total_high - loan_amount
interest_low = total_low - loan_amount

print("=== Impact of Interest Rate Drop Analysis ===")
print(f"Loan Amount: ${loan_amount:,.2f}")
print(f"Loan Term: {years} years ({months} payments)")
print()

print("SCENARIO 1: Original Rate (9.5%)")
print(f"  Monthly Payment: ${payment_high:,.2f}")
print(f"  Annual Payment: ${payment_high * 12:,.2f}")
print(f"  Total Payments: ${total_high:,.2f}")
print(f"  Total Interest: ${interest_high:,.2f}")
print(f"  Interest as % of loan: {(interest_high/loan_amount)*100:.1f}%")
print()

print("SCENARIO 2: Lower Rate (8.2%)")
print(f"  Monthly Payment: ${payment_low:,.2f}")
print(f"  Annual Payment: ${payment_low * 12:,.2f}")
print(f"  Total Payments: ${total_low:,.2f}")
print(f"  Total Interest: ${interest_low:,.2f}")
print(f"  Interest as % of loan: {(interest_low/loan_amount)*100:.1f}%")
print()

# Calculate savings
monthly_savings = payment_high - payment_low
annual_savings = monthly_savings * 12
total_interest_savings = interest_high - interest_low

print("IMPACT OF RATE DROP:")
print(f"  Monthly payment reduction: ${monthly_savings:,.2f}")
print(f"  Annual payment reduction: ${annual_savings:,.2f}")
print(f"  Total interest savings: ${total_interest_savings:,.2f}")
print(f"  Percentage reduction in total cost: {(total_interest_savings/total_high)*100:.1f}%")

# Create amortization schedules
schedule_high = create_amortization_schedule(loan_amount, rate_high, years)
schedule_low = create_amortization_schedule(loan_amount, rate_low, years)

# Create annual summaries
def create_annual_summary(schedule, years):
    annual_summary = []
    for year in range(1, years + 1):
        year_payments = [payment for payment in schedule if payment['year'] == year]

        if year_payments:
            annual_principal = sum(payment['principal'] for payment in year_payments)
            annual_interest = sum(payment['interest'] for payment in year_payments)
            year_end_balance = year_payments[-1]['balance']

            annual_summary.append({
                'year': year,
                'principal': annual_principal,
                'interest': annual_interest,
                'balance': year_end_balance
            })

    return annual_summary

annual_high = create_annual_summary(schedule_high, years)
annual_low = create_annual_summary(schedule_low, years)

# Display comparison for selected years
comparison_years = [1, 5, 10, 15, 20, 25, 30]
print(f"\n=== Annual Payment Breakdown Comparison ===")
print("Year | Rate 9.5%                    | Rate 8.2%                    | Difference")
print("     | Principal | Interest       | Principal | Interest       | Principal | Interest")
print("-----|-----------|----------------|-----------|----------------|-----------|----------")

for year in comparison_years:
    idx = year - 1
    high_data = annual_high[idx]
    low_data = annual_low[idx]

    prin_diff = low_data['principal'] - high_data['principal']
    int_diff = high_data['interest'] - low_data['interest']  # Higher rate has more interest

    print(f"  {year:2d} | ${high_data['principal']:8,.0f} | ${high_data['interest']:13,.0f} | ${low_data['principal']:8,.0f} | ${low_data['interest']:13,.0f} | ${prin_diff:+8,.0f} | ${int_diff:+8,.0f}")

# Create comprehensive visualizations
fig = plt.figure(figsize=(18, 12))

# Create subplot layout
ax1 = plt.subplot(2, 3, 1)
ax2 = plt.subplot(2, 3, 2)
ax3 = plt.subplot(2, 3, 3)
ax4 = plt.subplot(2, 3, 4)
ax5 = plt.subplot(2, 3, 5)
ax6 = plt.subplot(2, 3, 6)

fig.suptitle('Impact of Interest Rate Drop: 9.5% vs 8.2% on $500K 30-Year Mortgage', fontsize=16, fontweight='bold')

# 1. Payment Comparison
categories = ['9.5% Rate', '8.2% Rate']
monthly_payments = [payment_high, payment_low]
colors = ['lightcoral', 'lightgreen']

bars1 = ax1.bar(categories, monthly_payments, color=colors, alpha=0.8)
ax1.set_ylabel('Monthly Payment ($)')
ax1.set_title('Monthly Payment Comparison')
ax1.grid(axis='y', alpha=0.3)

# Add value labels and savings
for bar, value in zip(bars1, monthly_payments):
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2., height + 50,
             f'${value:,.0f}', ha='center', va='bottom', fontweight='bold')

ax1.text(0.5, max(monthly_payments) * 0.5, f'Monthly Savings:\n${monthly_savings:,.0f}',
         ha='center', transform=ax1.transAxes,
         bbox=dict(boxstyle="round,pad=0.5", facecolor="yellow", alpha=0.8))

# 2. Total Interest Comparison
total_interests = [interest_high, interest_low]
bars2 = ax2.bar(categories, total_interests, color=['red', 'darkgreen'], alpha=0.7)
ax2.set_ylabel('Total Interest ($)')
ax2.set_title('Total Interest Paid Over 30 Years')
ax2.grid(axis='y', alpha=0.3)

for bar, value in zip(bars2, total_interests):
    height = bar.get_height()
    ax2.text(bar.get_x() + bar.get_width()/2., height + 20000,
             f'${value:,.0f}', ha='center', va='bottom', fontweight='bold')

ax2.text(0.5, max(total_interests) * 0.7, f'Interest Savings:\n${total_interest_savings:,.0f}',
         ha='center', transform=ax2.transAxes,
         bbox=dict(boxstyle="round,pad=0.5", facecolor="lightgreen", alpha=0.8))

# 3. Monthly Principal vs Interest - High Rate
months_array = np.array([payment['month'] for payment in schedule_high])
principal_high = np.array([payment['principal'] for payment in schedule_high])
interest_high_monthly = np.array([payment['interest'] for payment in schedule_high])

ax3.plot(months_array, principal_high, label='Principal', color='green', linewidth=2)
ax3.plot(months_array, interest_high_monthly, label='Interest', color='red', linewidth=2)
ax3.set_xlabel('Month')
ax3.set_ylabel('Payment Amount ($)')
ax3.set_title('9.5% Rate: Principal vs Interest')
ax3.legend()
ax3.grid(alpha=0.3)

# 4. Monthly Principal vs Interest - Low Rate
principal_low = np.array([payment['principal'] for payment in schedule_low])
interest_low_monthly = np.array([payment['interest'] for payment in schedule_low])

ax4.plot(months_array, principal_low, label='Principal', color='green', linewidth=2)
ax4.plot(months_array, interest_low_monthly, label='Interest', color='red', linewidth=2)
ax4.set_xlabel('Month')
ax4.set_ylabel('Payment Amount ($)')
ax4.set_title('8.2% Rate: Principal vs Interest')
ax4.legend()
ax4.grid(alpha=0.3)

# 5. Balance Comparison Over Time
balances_high = np.array([payment['balance'] for payment in schedule_high])
balances_low = np.array([payment['balance'] for payment in schedule_low])

ax5.plot(months_array, balances_high, label='9.5% Rate', color='red', linewidth=3)
ax5.plot(months_array, balances_low, label='8.2% Rate', color='green', linewidth=3)
ax5.set_xlabel('Month')
ax5.set_ylabel('Remaining Balance ($)')
ax5.set_title('Mortgage Balance Over Time')
ax5.legend()
ax5.grid(alpha=0.3)

# Add annotation showing faster principal paydown
mid_point = months // 2
balance_diff_mid = balances_high[mid_point] - balances_low[mid_point]
ax5.annotate(f'Balance difference\nat 15 years:\n${balance_diff_mid:,.0f}',
             xy=(mid_point, balances_low[mid_point]),
             xytext=(mid_point + 50, balances_low[mid_point] + 50000),
             arrowprops=dict(arrowstyle='->', color='blue', alpha=0.7),
             bbox=dict(boxstyle="round,pad=0.3", facecolor="lightblue", alpha=0.8))

# 6. Annual Principal Payment Comparison
years_array = [item['year'] for item in annual_high]
annual_principal_high = [item['principal'] for item in annual_high]
annual_principal_low = [item['principal'] for item in annual_low]

width = 0.35
x_pos = np.arange(len(years_array))

bars_high = ax6.bar(x_pos - width/2, annual_principal_high, width, label='9.5% Rate', color='lightcoral', alpha=0.8)
bars_low = ax6.bar(x_pos + width/2, annual_principal_low, width, label='8.2% Rate', color='lightgreen', alpha=0.8)

ax6.set_xlabel('Year')
ax6.set_ylabel('Annual Principal Payment ($)')
ax6.set_title('Annual Principal Payments Comparison')
ax6.set_xticks(x_pos[::5])  # Show every 5th year
ax6.set_xticklabels([str(year) for year in years_array[::5]])
ax6.legend()
ax6.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# Additional Analysis
print(f"\n=== Detailed Impact Analysis ===")

# Find crossover points (when principal > interest)
crossover_high = None
crossover_low = None

for i, (p_h, int_h) in enumerate(zip(principal_high, interest_high_monthly)):
    if p_h > int_h and crossover_high is None:
        crossover_high = i + 1
        break

for i, (p_l, int_l) in enumerate(zip(principal_low, interest_low_monthly)):
    if p_l > int_l and crossover_low is None:
        crossover_low = i + 1
        break

if crossover_high and crossover_low:
    print(f"Principal exceeds interest payments:")
    print(f"  At 9.5% rate: Month {crossover_high} ({crossover_high/12:.1f} years)")
    print(f"  At 8.2% rate: Month {crossover_low} ({crossover_low/12:.1f} years)")
    print(f"  Difference: {(crossover_high - crossover_low)/12:.1f} years earlier with lower rate")

# Equity building comparison
print(f"\nEquity building comparison:")
milestones = [5, 10, 15, 20, 25]
for milestone in milestones:
    year_idx = milestone - 1
    equity_high = loan_amount - annual_high[year_idx]['balance']
    equity_low = loan_amount - annual_low[year_idx]['balance']
    equity_diff = equity_low - equity_high

    print(f"  After {milestone} years:")
    print(f"    9.5% rate equity: ${equity_high:,.0f}")
    print(f"    8.2% rate equity: ${equity_low:,.0f}")
    print(f"    Additional equity with lower rate: ${equity_diff:,.0f}")

# Break-even analysis for refinancing
refinance_cost = 5000  # Assume $5,000 refinancing cost
breakeven_months = refinance_cost / monthly_savings

print(f"\n=== Refinancing Analysis ===")
print(f"Assumed refinancing cost: ${refinance_cost:,.2f}")
print(f"Monthly savings: ${monthly_savings:,.2f}")
print(f"Break-even point: {breakeven_months:.1f} months ({breakeven_months/12:.1f} years)")
print(f"Net savings after break-even: ${total_interest_savings - refinance_cost:,.2f}")

if breakeven_months < 24:
    recommendation = "STRONGLY RECOMMENDED"
elif breakeven_months < 60:
    recommendation = "RECOMMENDED"
else:
    recommendation = "CONSIDER CAREFULLY"

print(f"\nRefinancing recommendation: {recommendation}")

# Sensitivity analysis
print(f"\n=== Sensitivity Analysis ===")
print("Impact of rate drop on different loan amounts:")
loan_amounts = [300000, 400000, 500000, 600000, 750000, 1000000]

for amount in loan_amounts:
    high_payment = calculate_monthly_payment(amount, rate_high, years)
    low_payment = calculate_monthly_payment(amount, rate_low, years)
    savings = high_payment - low_payment
    annual_savings = savings * 12

    print(f"  ${amount:,.0f} loan: ${savings:,.0f}/month (${annual_savings:,.0f}/year) savings")
````

````
#Scenario 12 & 13
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
import warnings
warnings.filterwarnings('ignore')

# Bond pricing function
def calculate_bond_price(face_value, coupon_rate, yield_rate, years_to_maturity, payments_per_year=1):
    """Calculate bond price given parameters"""
    periodic_coupon = (coupon_rate * face_value) / payments_per_year
    periodic_yield = yield_rate / payments_per_year
    total_periods = years_to_maturity * payments_per_year

    if periodic_yield == 0:
        # Handle zero yield case
        return face_value + (periodic_coupon * total_periods)

    # Present value of coupon payments
    pv_coupons = periodic_coupon * (1 - (1 + periodic_yield)**(-total_periods)) / periodic_yield

    # Present value of face value
    pv_face_value = face_value / (1 + periodic_yield)**total_periods

    return pv_coupons + pv_face_value

print("SCENARIO 12: Bond Price and Yield Relationship")
print("=" * 55)

# Base bond parameters for analysis
face_value = 1000
coupon_rate = 0.06  # 6%
years_to_maturity = 10
payments_per_year = 2  # Semi-annual

# Create yield range for plotting
yields = np.linspace(0.01, 0.12, 100)  # 1% to 12%
bond_prices = [calculate_bond_price(face_value, coupon_rate, y, years_to_maturity, payments_per_year)
               for y in yields]

# Create the main price-yield relationship plot
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))

# Plot 1: Classic Price-Yield Relationship
ax1.plot(yields * 100, bond_prices, 'b-', linewidth=2, label='Bond Price')
ax1.axhline(y=face_value, color='red', linestyle='--', alpha=0.7, label='Face Value ($1,000)')
ax1.axvline(x=coupon_rate * 100, color='green', linestyle='--', alpha=0.7, label='Coupon Rate (6%)')
ax1.set_xlabel('Yield to Maturity (%)')
ax1.set_ylabel('Bond Price ($)')
ax1.set_title('Bond Price vs Yield to Maturity\n(Inverse Relationship)')
ax1.grid(True, alpha=0.3)
ax1.legend()

# Add annotations for key points
par_price_index = np.argmin(np.abs(np.array(yields) - coupon_rate))
ax1.annotate(f'Par Value\n(${bond_prices[par_price_index]:.0f})',
             xy=(coupon_rate * 100, bond_prices[par_price_index]),
             xytext=(coupon_rate * 100 + 1, bond_prices[par_price_index] + 100),
             arrowprops=dict(arrowstyle='->', color='black'),
             fontsize=10, ha='center')

# Plot 2: Duration and Convexity Illustration
# Calculate duration approximation
duration_yields = np.array([0.05, 0.06, 0.07])
duration_prices = [calculate_bond_price(face_value, coupon_rate, y, years_to_maturity, payments_per_year)
                   for y in duration_yields]

ax2.plot(yields * 100, bond_prices, 'b-', linewidth=2, label='Actual Price')
ax2.plot(duration_yields * 100, duration_prices, 'ro', markersize=8, label='Sample Points')

# Linear approximation (duration)
center_idx = 1
slope = (duration_prices[2] - duration_prices[0]) / ((duration_yields[2] - duration_yields[0]) * 100)
intercept = duration_prices[center_idx] - slope * (duration_yields[center_idx] * 100)
linear_approx = slope * yields * 100 + intercept

ax2.plot(yields * 100, linear_approx, 'g--', alpha=0.7, label='Linear Approx (Duration)')
ax2.set_xlabel('Yield to Maturity (%)')
ax2.set_ylabel('Bond Price ($)')
ax2.set_title('Duration vs Convexity\n(Linear vs Curved Relationship)')
ax2.grid(True, alpha=0.3)
ax2.legend()
ax2.set_xlim(4, 8)

print("Key Observations from Price-Yield Relationship:")
print("-" * 45)
print("1. INVERSE RELATIONSHIP: As yield increases, price decreases")
print("2. CONVEX SHAPE: Price sensitivity is higher at lower yields")
print("3. PAR VALUE: When coupon rate = yield, bond trades at par")
print("4. PREMIUM/DISCOUNT: Coupon rate vs yield determines trading level")
print()

print("SCENARIO 13: Instant Experiments on Bond Price")
print("=" * 50)

# Base parameters for experiments
base_params = {
    'face_value': 1000,
    'coupon_rate': 0.06,
    'yield_rate': 0.08,
    'years_to_maturity': 10,
    'payments_per_year': 2
}

base_price = calculate_bond_price(**base_params)
print(f"Base Bond Price: ${base_price:.2f}")
print(f"Base Parameters: Face=${base_params['face_value']}, Coupon={base_params['coupon_rate']:.1%}, "
      f"Yield={base_params['yield_rate']:.1%}, Maturity={base_params['years_to_maturity']} years, "
      f"Payments/Year={base_params['payments_per_year']}")
print()

# Experiment (a): What happens when annual coupon rate is increased?
print("(a) Effect of Increasing Annual Coupon Rate:")
print("-" * 42)
coupon_rates = np.array([0.04, 0.06, 0.08, 0.10, 0.12])
coupon_prices = []
for rate in coupon_rates:
    params = base_params.copy()
    params['coupon_rate'] = rate
    price = calculate_bond_price(**params)
    coupon_prices.append(price)
    print(f"Coupon Rate: {rate:.1%} → Price: ${price:.2f} (Change: ${price - base_price:+.2f})")

ax3.plot(coupon_rates * 100, coupon_prices, 'bo-', linewidth=2, markersize=6)
ax3.axhline(y=base_price, color='red', linestyle='--', alpha=0.7, label='Base Price')
ax3.set_xlabel('Annual Coupon Rate (%)')
ax3.set_ylabel('Bond Price ($)')
ax3.set_title('Effect of Coupon Rate on Bond Price\n(Positive Relationship)')
ax3.grid(True, alpha=0.3)
ax3.legend()

print("\n→ RESULT: Higher coupon rate = Higher bond price (more cash flows)")
print()

# Experiment (b): What happens when yield to maturity is increased?
print("(b) Effect of Increasing Yield to Maturity:")
print("-" * 40)
ytm_rates = np.array([0.04, 0.06, 0.08, 0.10, 0.12])
ytm_prices = []
for rate in ytm_rates:
    params = base_params.copy()
    params['yield_rate'] = rate
    price = calculate_bond_price(**params)
    ytm_prices.append(price)
    print(f"YTM: {rate:.1%} → Price: ${price:.2f} (Change: ${price - base_price:+.2f})")

ax4.plot(ytm_rates * 100, ytm_prices, 'ro-', linewidth=2, markersize=6)
ax4.axhline(y=base_price, color='blue', linestyle='--', alpha=0.7, label='Base Price')
ax4.set_xlabel('Yield to Maturity (%)')
ax4.set_ylabel('Bond Price ($)')
ax4.set_title('Effect of Yield on Bond Price\n(Inverse Relationship)')
ax4.grid(True, alpha=0.3)
ax4.legend()

print("\n→ RESULT: Higher yield = Lower bond price (higher discount rate)")
print()

plt.tight_layout()
plt.show()

# Continue with more experiments
print("(c) Effect of Increasing Payments Per Year:")
print("-" * 38)
payment_frequencies = np.array([1, 2, 4, 12])  # Annual, Semi-annual, Quarterly, Monthly
payment_labels = ['Annual', 'Semi-annual', 'Quarterly', 'Monthly']
payment_prices = []
for freq in payment_frequencies:
    params = base_params.copy()
    params['payments_per_year'] = freq
    price = calculate_bond_price(**params)
    payment_prices.append(price)

for i, (freq, label, price) in enumerate(zip(payment_frequencies, payment_labels, payment_prices)):
    print(f"{label} ({freq}/year): ${price:.2f} (Change: ${price - base_price:+.2f})")

print("\n→ RESULT: More frequent payments = Slightly higher bond price (compounding effect)")
print()

# Experiment (d): What happens when face value is increased?
print("(d) Effect of Increasing Face Value:")
print("-" * 32)
face_values = np.array([500, 1000, 1500, 2000])
face_prices = []
for fv in face_values:
    params = base_params.copy()
    params['face_value'] = fv
    price = calculate_bond_price(**params)
    face_prices.append(price)
    print(f"Face Value: ${fv:,} → Price: ${price:.2f} (Ratio: {price/fv:.4f})")

print("\n→ RESULT: Bond price scales proportionally with face value")
print()

# Experiment (e): Relationship between par bond price and time to maturity
print("(e) Par Bond Price vs Time to Maturity:")
print("-" * 38)
maturities = np.array([1, 5, 10, 15, 20, 30])
par_prices = []
for maturity in maturities:
    params = base_params.copy()
    params['years_to_maturity'] = maturity
    params['yield_rate'] = base_params['coupon_rate']  # Make it a par bond
    price = calculate_bond_price(**params)
    par_prices.append(price)
    print(f"Maturity: {maturity:2d} years → Price: ${price:.2f}")

print("\n→ RESULT: Par bonds (coupon rate = yield) always trade at face value regardless of maturity")
print()

# Experiment (f): Coupon rate vs yield analysis
print("(f) Coupon Rate vs Yield Analysis:")
print("-" * 34)
test_yields = [0.04, 0.06, 0.08]
test_coupons = [0.04, 0.06, 0.08, 0.10]

print("Yield → ", end="")
for y in test_yields:
    print(f"{y:.1%}".rjust(10), end="")
print()
print("-" * 45)

for coupon in test_coupons:
    print(f"Coupon {coupon:.1%} → ", end="")
    for yield_rate in test_yields:
        params = base_params.copy()
        params['coupon_rate'] = coupon
        params['yield_rate'] = yield_rate
        price = calculate_bond_price(**params)
        if coupon == yield_rate:
            print(f"${price:.0f}*".rjust(10), end="")  # Mark par bonds
        else:
            print(f"${price:.0f}".rjust(10), end="")
    print()

print("\n* = Par bond (coupon rate = yield)")
print("\n→ RESULTS:")
print("  • When coupon rate = yield → Bond trades at PAR ($1,000)")
print("  • When coupon rate > yield → Bond trades at PREMIUM (> $1,000)")
print("  • When coupon rate < yield → Bond trades at DISCOUNT (< $1,000)")
print("  • Higher coupon rate → Higher price (for same yield)")
print()

# Create additional visualization for experiments
fig2, ((ax5, ax6), (ax7, ax8)) = plt.subplots(2, 2, figsize=(16, 12))

# Payment frequency effect
ax5.bar(payment_labels, payment_prices, alpha=0.7, color='lightgreen', edgecolor='black')
ax5.set_ylabel('Bond Price ($)')
ax5.set_title('Effect of Payment Frequency')
ax5.grid(True, alpha=0.3, axis='y')
for i, price in enumerate(payment_prices):
    ax5.text(i, price + 5, f'${price:.0f}', ha='center', fontweight='bold')

# Face value scaling
ax6.plot(face_values, face_prices, 'go-', linewidth=2, markersize=8)
ax6.set_xlabel('Face Value ($)')
ax6.set_ylabel('Bond Price ($)')
ax6.set_title('Bond Price Scaling with Face Value')
ax6.grid(True, alpha=0.3)

# Par bond maturity analysis
ax7.bar(range(len(maturities)), par_prices, alpha=0.7, color='gold', edgecolor='black')
ax7.set_xlabel('Years to Maturity')
ax7.set_ylabel('Bond Price ($)')
ax7.set_title('Par Bond Price vs Maturity\n(Always $1,000 when Coupon Rate = Yield)')
ax7.set_xticks(range(len(maturities)))
ax7.set_xticklabels(maturities)
ax7.grid(True, alpha=0.3, axis='y')
ax7.set_ylim(990, 1010)

# Heatmap for coupon vs yield
coupon_yield_matrix = np.zeros((len(test_coupons), len(test_yields)))
for i, coupon in enumerate(test_coupons):
    for j, yield_rate in enumerate(test_yields):
        params = base_params.copy()
        params['coupon_rate'] = coupon
        params['yield_rate'] = yield_rate
        price = calculate_bond_price(**params)
        coupon_yield_matrix[i, j] = price

im = ax8.imshow(coupon_yield_matrix, cmap='RdYlGn', aspect='auto')
ax8.set_xticks(range(len(test_yields)))
ax8.set_xticklabels([f'{y:.1%}' for y in test_yields])
ax8.set_yticks(range(len(test_coupons)))
ax8.set_yticklabels([f'{c:.1%}' for c in test_coupons])
ax8.set_xlabel('Yield to Maturity')
ax8.set_ylabel('Coupon Rate')
ax8.set_title('Bond Price Heatmap\n(Green=Premium, Red=Discount)')

# Add text annotations to heatmap
for i in range(len(test_coupons)):
    for j in range(len(test_yields)):
        text = ax8.text(j, i, f'${coupon_yield_matrix[i, j]:.0f}',
                       ha="center", va="center", color="black", fontweight='bold')

plt.colorbar(im, ax=ax8, label='Bond Price ($)')
plt.tight_layout()
plt.show()

print("SUMMARY OF KEY BOND PRICING PRINCIPLES:")
print("=" * 40)
print("1. PRICE-YIELD RELATIONSHIP: Always inverse and convex")
print("2. COUPON EFFECT: Higher coupons → Higher prices")
print("3. MATURITY EFFECT: Longer maturity → Higher price sensitivity")
print("4. FREQUENCY EFFECT: More payments → Slightly higher prices")
print("5. PAR BOND RULE: Coupon rate = Yield → Price = Face value")
print("6. PREMIUM/DISCOUNT: Determined by coupon rate vs yield comparison")
print("7. CONVEXITY: Price increases more when yields fall than decreases when yields rise")
````

````
#Scenario 14
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar

# Bond parameters
face_value = 1000
annual_coupon_rate = 0.045  # 4.5%
yield_to_maturity = 0.035  # 3.5%
years_to_maturity = 6
payments_per_year = 2  # Semi-annual
total_periods = years_to_maturity * payments_per_year  # 12 periods

print("DURATION AND CONVEXITY ANALYSIS")
print("=" * 50)
print(f"Bond Parameters:")
print(f"Face Value: ${face_value:,}")
print(f"Annual Coupon Rate: {annual_coupon_rate:.1%}")
print(f"Yield to Maturity: {yield_to_maturity:.1%}")
print(f"Years to Maturity: {years_to_maturity}")
print(f"Payment Frequency: {payments_per_year} times per year")
print(f"Total Periods: {total_periods}")
print()

# Semi-annual coupon payment
semi_annual_coupon = (annual_coupon_rate * face_value) / payments_per_year
print(f"Semi-annual Coupon Payment: ${semi_annual_coupon:.2f}")
print()

def calculate_bond_metrics_apr(face_value, annual_coupon, ytm, years, freq):
    """Calculate bond price, duration, and convexity using APR convention"""
    # APR: Simple division
    periodic_rate = ytm / freq
    total_periods = years * freq
    periodic_coupon = annual_coupon / freq

    # Time periods
    periods = np.arange(1, total_periods + 1)
    times_in_years = periods / freq

    # Cash flows (coupon payments + face value at maturity)
    cash_flows = np.full(total_periods, periodic_coupon)
    cash_flows[-1] += face_value  # Add face value to last payment

    # Present values
    discount_factors = (1 + periodic_rate) ** (-periods)
    present_values = cash_flows * discount_factors

    # Bond price
    bond_price = np.sum(present_values)

    # Macaulay Duration (in years)
    weighted_times = times_in_years * present_values
    macaulay_duration = np.sum(weighted_times) / bond_price

    # Modified Duration
    modified_duration = macaulay_duration / (1 + periodic_rate)

    # Convexity calculation
    # Convexity = Σ[t(t+1) * PV(CF_t)] / [P * (1+r)^2]
    convexity_weights = periods * (periods + 1) * present_values / (freq ** 2)
    convexity = np.sum(convexity_weights) / (bond_price * (1 + periodic_rate) ** 2)

    return {
        'bond_price': bond_price,
        'macaulay_duration': macaulay_duration,
        'modified_duration': modified_duration,
        'convexity': convexity,
        'periodic_rate': periodic_rate,
        'cash_flows': cash_flows,
        'present_values': present_values,
        'periods': periods,
        'times_in_years': times_in_years
    }

def calculate_bond_metrics_ear(face_value, annual_coupon, ytm, years, freq):
    """Calculate bond price, duration, and convexity using EAR convention"""
    # EAR: True equivalent rate
    periodic_rate = (1 + ytm) ** (1/freq) - 1
    total_periods = years * freq
    periodic_coupon = annual_coupon / freq

    # Time periods
    periods = np.arange(1, total_periods + 1)
    times_in_years = periods / freq

    # Cash flows
    cash_flows = np.full(total_periods, periodic_coupon)
    cash_flows[-1] += face_value

    # Present values
    discount_factors = (1 + periodic_rate) ** (-periods)
    present_values = cash_flows * discount_factors

    # Bond price
    bond_price = np.sum(present_values)

    # Macaulay Duration (in years)
    weighted_times = times_in_years * present_values
    macaulay_duration = np.sum(weighted_times) / bond_price

    # Modified Duration
    modified_duration = macaulay_duration / (1 + periodic_rate)

    # Convexity
    convexity_weights = periods * (periods + 1) * present_values / (freq ** 2)
    convexity = np.sum(convexity_weights) / (bond_price * (1 + periodic_rate) ** 2)

    return {
        'bond_price': bond_price,
        'macaulay_duration': macaulay_duration,
        'modified_duration': modified_duration,
        'convexity': convexity,
        'periodic_rate': periodic_rate,
        'cash_flows': cash_flows,
        'present_values': present_values,
        'periods': periods,
        'times_in_years': times_in_years
    }

# Calculate using both methods
print("CALCULATIONS USING APR CONVENTION")
print("=" * 40)
apr_results = calculate_bond_metrics_apr(face_value, annual_coupon_rate * face_value,
                                        yield_to_maturity, years_to_maturity, payments_per_year)

print(f"Periodic Rate (APR): {apr_results['periodic_rate']:.4%}")
print(f"Bond Price: ${apr_results['bond_price']:.2f}")
print(f"Macaulay Duration: {apr_results['macaulay_duration']:.4f} years")
print(f"Modified Duration: {apr_results['modified_duration']:.4f} years")
print(f"Convexity: {apr_results['convexity']:.4f}")
print()

print("CALCULATIONS USING EAR CONVENTION")
print("=" * 40)
ear_results = calculate_bond_metrics_ear(face_value, annual_coupon_rate * face_value,
                                        yield_to_maturity, years_to_maturity, payments_per_year)

print(f"Periodic Rate (EAR): {ear_results['periodic_rate']:.4%}")
print(f"Bond Price: ${ear_results['bond_price']:.2f}")
print(f"Macaulay Duration: {ear_results['macaulay_duration']:.4f} years")
print(f"Modified Duration: {ear_results['modified_duration']:.4f} years")
print(f"Convexity: {ear_results['convexity']:.4f}")
print()

# Create detailed cash flow analysis
cash_flow_data = {
    'Period': apr_results['periods'],
    'Time (Years)': apr_results['times_in_years'],
    'Cash Flow': apr_results['cash_flows'],
    'PV (APR)': apr_results['present_values'],
    'PV (EAR)': ear_results['present_values'],
    'Time × PV (APR)': apr_results['times_in_years'] * apr_results['present_values'],
    'Time × PV (EAR)': ear_results['times_in_years'] * ear_results['present_values']
}

df = pd.DataFrame(cash_flow_data)
print("DETAILED CASH FLOW AND DURATION ANALYSIS")
print("=" * 50)
print(df.to_string(index=False, float_format='%.2f'))
print()

# Comparison table
comparison_data = {
    'Metric': ['Bond Price ($)', 'Macaulay Duration (years)', 'Modified Duration (years)', 'Convexity'],
    'APR Convention': [
        f"${apr_results['bond_price']:.2f}",
        f"{apr_results['macaulay_duration']:.4f}",
        f"{apr_results['modified_duration']:.4f}",
        f"{apr_results['convexity']:.4f}"
    ],
    'EAR Convention': [
        f"${ear_results['bond_price']:.2f}",
        f"{ear_results['macaulay_duration']:.4f}",
        f"{ear_results['modified_duration']:.4f}",
        f"{ear_results['convexity']:.4f}"
    ]
}

comparison_df = pd.DataFrame(comparison_data)
print("COMPARISON: APR vs EAR CONVENTIONS")
print("=" * 40)
print(comparison_df.to_string(index=False))
print()

# Duration approximation test
def test_duration_approximation(base_results, ytm_change=0.01, convention='APR'):
    """Test how well duration approximates price changes"""
    new_ytm = yield_to_maturity + ytm_change

    if convention == 'APR':
        new_results = calculate_bond_metrics_apr(face_value, annual_coupon_rate * face_value,
                                               new_ytm, years_to_maturity, payments_per_year)
    else:
        new_results = calculate_bond_metrics_ear(face_value, annual_coupon_rate * face_value,
                                               new_ytm, years_to_maturity, payments_per_year)

    # Actual price change
    actual_price_change = new_results['bond_price'] - base_results['bond_price']
    actual_percentage_change = actual_price_change / base_results['bond_price']

    # Duration approximation
    duration_approximation = -base_results['modified_duration'] * ytm_change

    # Duration + Convexity approximation
    convexity_adjustment = 0.5 * base_results['convexity'] * (ytm_change ** 2)
    duration_convexity_approximation = duration_approximation + convexity_adjustment

    return {
        'actual_change': actual_percentage_change,
        'duration_approx': duration_approximation,
        'duration_convexity_approx': duration_convexity_approximation,
        'duration_error': abs(actual_percentage_change - duration_approximation),
        'convexity_error': abs(actual_percentage_change - duration_convexity_approximation)
    }

# Test duration approximations
yield_changes = np.array([-0.02, -0.01, -0.005, 0.005, 0.01, 0.02])
print("DURATION APPROXIMATION ACCURACY TEST")
print("=" * 45)
print("YTM Change | Actual Δ% | Duration | Dur+Conv | Dur Error | Conv Error")
print("-" * 70)

for dy in yield_changes:
    test_apr = test_duration_approximation(apr_results, dy, 'APR')
    print(f"{dy:+8.3%} | {test_apr['actual_change']:+8.4%} | {test_apr['duration_approx']:+8.4%} | "
          f"{test_apr['duration_convexity_approx']:+8.4%} | {test_apr['duration_error']:8.4%} | {test_apr['convexity_error']:8.4%}")

print()
print("→ OBSERVATION: Duration alone works well for small yield changes")
print("→ OBSERVATION: Convexity adjustment improves accuracy for larger changes")
print()

# Visualizations
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))

# Plot 1: Cash Flow Timeline
periods_years = apr_results['times_in_years']
cash_flows = apr_results['cash_flows']

bars = ax1.bar(periods_years, cash_flows, alpha=0.7, color='lightblue', edgecolor='navy', width=0.2)
ax1.set_xlabel('Time (Years)')
ax1.set_ylabel('Cash Flow ($)')
ax1.set_title('Semi-Annual Cash Flow Timeline')
ax1.grid(True, alpha=0.3)

# Highlight the final payment
ax1.annotate(f'${cash_flows[-1]:.0f}\n(${semi_annual_coupon:.0f} + ${face_value:,})',
             xy=(periods_years[-1], cash_flows[-1]),
             xytext=(periods_years[-1]-0.5, cash_flows[-1]+50),
             arrowprops=dict(arrowstyle='->', color='red'),
             ha='center', fontsize=10)

# Plot 2: Present Value Weights for Duration
pv_weights_apr = apr_results['present_values'] / apr_results['bond_price']
pv_weights_ear = ear_results['present_values'] / ear_results['bond_price']

width = 0.15
x_pos = np.arange(len(periods_years))
ax2.bar(x_pos - width/2, pv_weights_apr, width, label='APR', alpha=0.7, color='lightcoral')
ax2.bar(x_pos + width/2, pv_weights_ear, width, label='EAR', alpha=0.7, color='lightgreen')
ax2.set_xlabel('Period')
ax2.set_ylabel('Present Value Weight')
ax2.set_title('Present Value Weights by Period\n(Higher weights = more duration impact)')
ax2.set_xticks(x_pos)
ax2.set_xticklabels([f'{t:.1f}' for t in periods_years])
ax2.legend()
ax2.grid(True, alpha=0.3)

# Plot 3: Duration Approximation Accuracy
actual_changes = []
duration_approx = []
duration_convexity_approx = []

for dy in yield_changes:
    test = test_duration_approximation(apr_results, dy, 'APR')
    actual_changes.append(test['actual_change'] * 100)
    duration_approx.append(test['duration_approx'] * 100)
    duration_convexity_approx.append(test['duration_convexity_approx'] * 100)

ax3.plot(yield_changes * 100, actual_changes, 'bo-', label='Actual Price Change', linewidth=2, markersize=6)
ax3.plot(yield_changes * 100, duration_approx, 'r--', label='Duration Approximation', linewidth=2)
ax3.plot(yield_changes * 100, duration_convexity_approx, 'g:', label='Duration + Convexity', linewidth=2)
ax3.set_xlabel('Yield Change (%)')
ax3.set_ylabel('Price Change (%)')
ax3.set_title('Duration Approximation Accuracy')
ax3.legend()
ax3.grid(True, alpha=0.3)

# Plot 4: Price-Yield Curve with Duration Tangent
yield_range = np.linspace(0.01, 0.06, 50)
prices_apr = []
for y in yield_range:
    result = calculate_bond_metrics_apr(face_value, annual_coupon_rate * face_value,
                                      y, years_to_maturity, payments_per_year)
    prices_apr.append(result['bond_price'])

ax4.plot(yield_range * 100, prices_apr, 'b-', linewidth=2, label='Actual Price Curve')
ax4.plot(yield_to_maturity * 100, apr_results['bond_price'], 'ro', markersize=8, label='Current Point')

# Duration tangent line
base_ytm = yield_to_maturity * 100
base_price = apr_results['bond_price']
slope = -apr_results['modified_duration'] * base_price / 100  # Convert to price per % yield change

tangent_yields = np.array([base_ytm - 1, base_ytm + 1])
tangent_prices = base_price + slope * (tangent_yields - base_ytm)
ax4.plot(tangent_yields, tangent_prices, 'g--', linewidth=2, label='Duration Tangent', alpha=0.7)

ax4.set_xlabel('Yield to Maturity (%)')
ax4.set_ylabel('Bond Price ($)')
ax4.set_title('Price-Yield Curve with Duration Tangent')
ax4.legend()
ax4.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Intuitive interpretation
print("INTUITIVE INTERPRETATION OF DURATION")
print("=" * 40)
print(f"📊 MACAULAY DURATION: {apr_results['macaulay_duration']:.2f} years")
print("   → This is the 'weighted average time' to receive the bond's cash flows")
print("   → Think of it as the bond's 'center of gravity' in time")
print("   → Earlier cash flows reduce duration, later cash flows increase it")
print()
print(f"⚡ MODIFIED DURATION: {apr_results['modified_duration']:.2f} years")
print("   → This measures price sensitivity to yield changes")
print(f"   → A 1% increase in yield causes ~{apr_results['modified_duration']:.2f}% price decrease")
print(f"   → A 1% decrease in yield causes ~{apr_results['modified_duration']:.2f}% price increase")
print("   → It's the slope of the price-yield curve (approximately)")
print()
print(f"🏹 CONVEXITY: {apr_results['convexity']:.2f}")
print("   → This measures how duration itself changes with yield changes")
print("   → Positive convexity is GOOD for bondholders:")
print("   →   • Prices rise MORE when yields fall")
print("   →   • Prices fall LESS when yields rise")
print("   → It's the curvature of the price-yield relationship")
print()

print("PRACTICAL APPLICATIONS:")
print("=" * 25)
print("1. PORTFOLIO RISK MANAGEMENT:")
print(f"   → This bond has {apr_results['modified_duration']:.1f} years of interest rate risk")
print("   → If rates rise 1%, expect ~{:.1f}% loss in bond value".format(apr_results['modified_duration']))
print()
print("2. HEDGING STRATEGIES:")
print("   → To hedge $1M of this bond, you'd need securities with opposite duration")
print("   → Duration matching helps immunize portfolios against rate changes")
print()
print("3. ACTIVE BOND MANAGEMENT:")
print("   → When expecting rates to fall → buy high duration bonds")
print("   → When expecting rates to rise → buy low duration bonds")
print("   → Convexity provides additional upside when rates are volatile")
print()

print("KEY DIFFERENCES: APR vs EAR")
print("=" * 30)
price_diff = ear_results['bond_price'] - apr_results['bond_price']
duration_diff = ear_results['modified_duration'] - apr_results['modified_duration']
convexity_diff = ear_results['convexity'] - apr_results['convexity']

print(f"Price difference: ${price_diff:.2f} ({price_diff/apr_results['bond_price']*100:+.3f}%)")
print(f"Duration difference: {duration_diff:+.4f} years")
print(f"Convexity difference: {convexity_diff:+.4f}")
print()
print("→ EAR convention typically produces slightly different results")
print("→ Differences are small but can be meaningful for large portfolios")
print("→ EAR is more theoretically correct for compounding analysis")
````
